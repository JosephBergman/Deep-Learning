## Step 2: Preparing the Data
As always, we have to prepare our data before feeding it to a neural network. Here are the steps:
1. Convert the images into 2D tensors with shape `(num_samples, dimension)`
2. Scale the images to take a range of values between 0 and 1
3. One-hot encode the labels and reshape as a 2D tensor with shape `(num_samples, num_classes)`
4. Split the data into training, validation, and test sets.

#### Step 1: Convert the images into 2D tensors
Fashion-MNIST really is a drop in replacement for MNIST, so it should be unsurprising that there are 60,000 training samples and 10,000 test samples. We'll double check this below, and then we'll reshape the images into 2D tensors.

```python
print('Training shape: ', training_data.shape)
print('Test shape: ', test_data.shape)
```

```python
training_data = training_data.reshape(60000, 28*28)
test_data = test_data.reshape(10000, 28*28)

print('Training shape: ', training_data.shape)
print('Test shape: ', test_data.shape)
```

#### Step 2: Scale the images to take values between 0 and 1
Currently, our grayscale images take values between 0 and 255. We will scale them to take values between 0 and 1 by dividing by the maximum value. First, we can double check that they haven't already been scaled, then we will scale them.

```python
print('Maximum pixel value', max(training_data[0]))
print('Minimum pixel value', min(training_data[0]))
```

```python
training_data = training_data.astype('float32') / 255
test_data = test_data.astype('float32') / 255

print('Maximum pixel value', max(training_data[0]))
print('Minimum pixel value', min(training_data[0]))
```

#### Step 3: One-hot encode our labels and reshape as a 2D tensor
Finally, our labels are again just integer values. We need to convert them to one-hot vectors that can be used with a softmax layer in a neural network. This is simple enough with the `to_categorial` function provided by Keras. Below, let's look at some of the raw labels, encode them, then take another look. If you are still confused by one-hot encoding, compare the raw training labels with the one-hot vectors.

```python
print(training_labels[:10])
```

```python
from keras.utils import to_categorical

training_labels = to_categorical(training_labels)
test_labels = to_categorical(test_labels)

print('Training label shape', training_labels.shape)
print('Testing label shape', test_labels.shape)
```

```python
print(training_labels[:10])
```

#### Step 4: Split the data into training, validation, and test sets
We will use the same 50,000/10,000/10,000 train/validation/test split we used with the MNIST example. After splitting the data we will have the following datasets:
+ `x_train, y_train` for iteratively developing several different models.
+ `x_val, y_val` for testing our various models on unseen data.
+ `training_data, training_labels` for training our final model on all our non-test data.
+ `test_data, test_labels` for sanity checking our final model on truly unseen data.

By the way, here is brief warning for future projects. We are taking the first 10,000 examples for our validation set. Sometimes datasets will be ordered, for example, the first 10,000 examples only contains boots and t-shirts, and the other eight classes are contained in the remaining 50,000 examples. If this is the case, our training and validation sets would represent completely different objects. When this is the case, you should randomly shuffle your dataset before creating a train/validation/test split. Fashion-MNIST is already shuffled, so we don't have to worry about it.

```python
x_val = training_data[:10000]
x_train = training_data[10000:]

y_val = training_labels[:10000]
y_train = training_labels[10000:]
```

## Step 3: Developing our Neural Network

In the last blog post, we introduced the following five step approach to developing our neural network
1. Load the data 
2. Prepare the data 
3. Iteratively develop our neural network
4. Train a final neural network using all training data
5. Evaluate our final network on the test set

Step 3 is usually quite long because of the number of hyperparameters to tune, especially for deep neural networks. In this post, we are going to spend extra time on step 3 by examining  different techniques we can use to improve the model in terms of training accuracy, learning speed, and amount of overfitting.

```python
import matplotlib.pyplot as plt 

def display_loss(history):
    """Display the training/validation loss from the history."""
    train_loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = range(1, len(train_loss) + 1)

    plt.plot(epochs, train_loss, 'g', label='Training Loss')
    plt.plot(epochs, val_loss, 'b', label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
    
def display_accuracy(history):
    """Display the train/validation accuracy from the history."""
    train_acc = history.history['acc']
    val_acc = history.history['val_acc']
    epochs = range(1, len(train_acc) + 1)
    
    plt.clf() 
    plt.plot(epochs, train_acc, 'g', label='Training Accurcy')
    plt.plot(epochs, val_acc, 'b', label='Validation Accurcy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

#### Establishing a Baseline Model
We'll start with a very simple model containing only a single hidden layer with just 16 units.

```python
from keras import models 
from keras import layers 
from keras import optimizers

# Define the architecture
network = models.Sequential() 
network.add(layers.Dense(16, activation='relu', input_shape=(x_train[0].shape)))
network.add(layers.Dense(10, activation='softmax'))

# Define the optimizer, loss, and metrics 
network.compile(optimizer='sgd',
                loss='categorical_crossentropy',
                metrics=['accuracy'])

# Train the network
history = network.fit(x_train, 
                      y_train,
                      epochs=20,
                      batch_size=256,
                      verbose=0,
                      validation_data=(x_val, y_val))

display_loss(history)
display_accuracy(history)
print("Training Accuracy: ", history.history['acc'][-1])
print("Validation Accuracy: ", history.history['val_acc'][-1])
```

Our training and validation accuracy are very similar – which is a good thing – but our training accuracy is quite low. Recognizing 28x28 images of fashion items is quite easy for humans, so we would hope that a neural network can do better than 82.8% on the task. In fact, doing well on the training set is essentially a prerequisite for building a good model. Let's create a network with one hidden layer and see if we can do better on the training set.

```python
from keras import models 
from keras import layers 

# Define the architecture
network = models.Sequential() 
network.add(layers.Dense(512, activation='relu', input_shape=(x_train[0].shape)))
network.add(layers.Dense(10, activation='softmax'))

# Define the optimizer, loss, and metrics 
network.compile(optimizer='adam',
                loss='categorical_crossentropy',
                metrics=['accuracy'])

# Train the network
history = network.fit(x_train, 
                      y_train,
                      epochs=100,
                      batch_size=256,
                      verbose=0,
                      validation_data=(x_val, y_val))

display_loss(history)
display_accuracy(history)
print("Training Accuracy: ", history.history['acc'][-1])
print("Validation Accuracy: ", history.history['val_acc'][-1])
```

Perfect! 99.1% training accuracy and 89.4% validation accuracy. The gap between the training accuracy and the validation accuracy is characteristic of overfitting and shows that our model has started to learn features unique to the training set – in other words, they don't 'generalize' to unseen examples. We will look at how to combat overfitting in the section on regularization.

## Initialization
When a neural network is first created, all the weights are randomly set. This is important, if all the weights were 0 or if the weights were symmetric, then the network may not be able to learn. However, you can't just randomly initialize the weights to any values. Random weights that are too large may make learning incredibly slow or impossible. In this section, we will look at three different initialization approaches and see how they compare. 

To get started, let's create a function which returns a network with a given number of hidden layers, hidden units, and the specified initializer.  

```python
from keras import models 
from keras import layers 

def initializer_model(num_nodes, num_layers, initializer):
    """Helper for creating models with different layer initializers.
    
    Arguments 
        num_nodes: The number of hidden units in each hidden layer 
        num_layers: The number of hidden layers in the network 
        initializer: The initializer to use for each kernel 
        
    Returns 
        A model with the specified parameters 
    """
    network = models.Sequential() 
    network.add(layers.Dense(num_nodes, activation='relu', kernel_initializer=initializer, input_shape=(x_train[0].shape)))
    for _ in range(num_layers-1):
        network.add(layers.Dense(num_nodes, activation='relu', kernel_initializer=initializer))
    network.add(layers.Dense(10, kernel_initializer=initializer, activation='softmax'))
    network.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
    return network
```

Next, we will import the initializers module and compare the three methods: `RandomUniform`, `glorot_uniform`, and `he_uniform`. We set the minimum value and maximum value of RandomUniform to -1 and 1, which is quite large compared to the -0.05 and 0.05 default values. This is primarily to make a point that weight initializations do matter. `glorot_uniform` (also known as a Xavier initializer) is the default used if you don't specify an initialization method. `he_uniform` is a He initializer. We will create three 1-layer, three 2-layer, and three 3-layer networks using each of the initializers.

```python
from keras import initializers

# Define our different initializers
random = initializers.RandomUniform(minval=-1, maxval=1)
xavier = initializers.glorot_uniform()
he = initializers.he_uniform()

# 1-layer networks with 32 nodes 
small32random = initializer_model(32, 1, random)
small32xavier = initializer_model(32, 1, xavier)
small32he = initializer_model(32, 1, he)

# 2-layer networks with 32 nodes 
medium32random = initializer_model(32, 2, random)
medium32xavier = initializer_model(32, 2, xavier)
medium32he = initializer_model(32, 2, he)

# 3-layer networks with 32 nodes 
large32random = initializer_model(32, 3, random)
large32xavier = initializer_model(32, 3, xavier)
large32he = initializer_model(32, 3, he)
```

We'll write some helper methods to train all of our models and compare their accuracy.

```python
def train_model(model):
    return model.fit(training_data, training_labels, 
                     epochs=30, batch_size=256, verbose=0)
    
def train_models(modelA, modelB, modelC):
    print("Training Model A...")
    modelA_history = train_model(modelA)
    print("Training Model B...")
    modelB_history = train_model(modelB)
    print("Training Model C...")
    modelC_history = train_model(modelC)
    return (modelA_history, modelB_history, modelC_history)

def display_models_loss(modelA, modelB, modelC, labels):
    modelA_loss = modelA.history['loss']
    modelB_loss = modelB.history['loss']
    modelC_loss = modelC.history['loss']
    epochs = range(1, len(modelA_loss) + 1)

    plt.plot(epochs, modelA_loss, 'r', label=labels[0])
    plt.plot(epochs, modelB_loss, 'b', label=labels[1])
    plt.plot(epochs, modelC_loss, 'g', label=labels[2])
    plt.title('Model Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
    
def display_models_accuracy(modelA, modelB, modelC, labels):
    modelA_acc = modelA.history['acc']
    modelB_acc = modelB.history['acc']
    modelC_acc = modelC.history['acc']
    epochs = range(1, len(modelA_acc) + 1)

    plt.plot(epochs, modelA_acc, 'r', label=labels[0])
    plt.plot(epochs, modelB_acc, 'b', label=labels[1])
    plt.plot(epochs, modelC_acc, 'g', label=labels[2])
    plt.title('Model Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
    
def print_models_accuracy(modelA, modelB, modelC, labels):
    print(labels[0], modelA.history['acc'][-1])
    print(labels[1], modelB.history['acc'][-1])
    print(labels[2], modelC.history['acc'][-1])
    
def compare_models(modelA, modelB, modelC, labels):
    display_models_loss(modelA, modelB, modelC, labels)
    display_models_accuracy(modelA, modelB, modelC, labels)
    print_models_accuracy(modelA, modelB, modelC, labels)
```

### Small Models
Let's start by training the small models and comparing the results.

```python
(small_random, small_xavier, small_he) = train_models(small32random, small32xavier, small32he)
```

```python
display_models_loss(small_random, small_xavier, small_he, ["Random Loss", "Xavier Loss", "He Loss"])
display_models_accuracy(small_random, small_xavier, small_he, ["Random Accuracy", "Xavier Accuracy", "He Accuracy"])

print('Small Random Accuracy: ', small_random.history['acc'][-1])
print('Small Xavier Accuracy: ', small_xavier.history['acc'][-1])
print('Small He Accuracy: ', small_he.history['acc'][-1])
```

We see that the He initializer does the best, barely beating the Xavier (`glorot_uniform`) initializer. The randomly initialized weights perform notably worse than the other two intializers showing that the magnitude of the initialized weights does matter.

### Medium Models
Let's repeat this experiment with the medium sized models.

```python
(med_random, med_xavier, med_he) = train_models(medium32random, medium32xavier, medium32he)
```

```python
display_models_loss(med_random, med_xavier, med_he, ["Random Loss", "Xavier Loss", "He Loss"])
display_models_accuracy(med_random, med_xavier, med_he, ["Random Accuracy", "Xavier Accuracy", "He Accuracy"])

print('Medium Random Accuracy: ', med_random.history['acc'][-1])
print('Medium Xavier Accuracy: ', med_xavier.history['acc'][-1])
print('Medium He Accuracy: ', med_he.history['acc'][-1])
```

Again, we see that He outperforms Xavier which outperforms random.

### Large Models
Finally, we will run the experiment for the 3-layer models.

```python
(large_random, large_xavier, large_he) = train_models(large32random, large32xavier, large32he)
```

```python
display_models_loss(large_random, large_xavier, large_he, ["Random Loss", "Xavier Loss", "He Loss"])
display_models_accuracy(large_random, large_xavier, large_he, ["Random Accuracy", "Xavier Accuracy", "He Accuracy"])

print('Large Random Accuracy: ', large_random.history['acc'][-1])
print('Large Xavier Accuracy: ', large_xavier.history['acc'][-1])
print('Large He Accuracy: ', large_he.history['acc'][-1])
```

The difference between Xavier and He is small in all cases, but in all cases He also had the slight edge.

## Optimizers 
In this section, we will compare three different optimizers: SGD, RMSprop, and Adam. Optimizers are the algorithms used during training to update the weights and biases in the network. They can have a significant impact on the accuracy and speed of training. In the last post, we used Stochastic Gradient Descent as it is the most basic of the optimizers. In this section, we'll see that it's generally better to use RMSprop or Adam. 

Similar to the previous section, we'll define a helper to generate our models.

```python
def optimizer_model(num_nodes, num_layers, optimizer):
    """Helper for creating models with different layer initializers.
    
    Arguments 
        num_nodes: The number of hidden units in each hidden layer 
        num_layers: The number of hidden layers in the network 
        initializer: The initializer to use for each kernel 
        
    Returns 
        A model with the specified parameters 
    """
    network = models.Sequential() 
    
    # Define the architecture
    network.add(layers.Dense(num_nodes, activation='relu', kernel_initializer='he_uniform', input_shape=(x_train[0].shape)))
    
    for _ in range(num_layers-1):
        network.add(layers.Dense(num_nodes, activation='relu', kernel_initializer='he_uniform'))
        
    network.add(layers.Dense(10, activation='softmax', kernel_initializer='he_uniform'))
    
    # Pass in the specified architecture
    network.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return network
```

The code below will create 9 different models. Three small models (1-layer with 32 hidden nodes), three medium models (2-layers with 32 hidden nodes layer), and three large models (3-layers with 32 hidden nodes per layer). Of course, a network with three layers is not large by modern standards, but small, medium, and large are good names to keep track of the different networks. For each network size, one model will use Stochastic Gradient Descent, one will use RMSProp, and one will use an Adam optimizer.

```python
# Define our different optimizers
sgd = optimizers.SGD()
rms = optimizers.RMSprop()
adam = optimizers.Adam()

# 1-layer networks with 32 nodes 
small32sgd = optimizer_model(32, 1, sgd)
small32rms = optimizer_model(32, 1, rms)
small32adam = optimizer_model(32, 1, adam)

# 2-layer networks with 32 nodes 
medium32sgd = optimizer_model(32, 2, sgd)
medium32rms = optimizer_model(32, 2, rms)
medium32adam = optimizer_model(32, 2, adam)

# 3-layer networks with 32 nodes 
large32sgd = optimizer_model(32, 3, sgd)
large32rms = optimizer_model(32, 3, rms)
large32adam = optimizer_model(32, 3, adam)
```

Once the models have been created, we have to train them. Expect this to take a couple of minutes.

```python
# Train small models 
(small_sgd, small_rms, small_adam) = train_models(small32sgd, small32rms, small32adam)

# Train medium models
(medium_sgd, medium_rms, medium_adam) = train_models(medium32sgd, medium32rms, medium32adam)

# Train large models
(large_sgd, large_rms, large_adam) = train_models(large32sgd, large32rms, large32adam)
```

#### Small Models

```python
compare_models(small_sgd, small_rms, small_adam, ['SGD', 'RMS', 'Adam'])
```

#### Medium Models

```python
compare_models(medium_sgd, medium_rms, medium_adam, ['SGD', 'RMS', 'Adam'])
```

#### Large Models

```python
compare_models(large_sgd, large_rms, large_adam, ['SGD', 'RMS', 'Adam'])
```

### Summary 
The results are briefly summarize in the following table. We can see that in all cases SGD performed the worst, then Adam, and RMSprop always performed the best. Generally speaking, RMSprop is usually a safe choice.

## Regularization
In this section, we are going to look at methods for addressing overfitting
+ L1 Regularization
+ L2 Regularization 
+ Dropout

In general, regularization works by punishing the network for weights that are too large. This forces the network to learn a simpler or more general representation, and helps to prevent overfitting on training-set, unique features. In the case of L1 regularization, weights are punished based on their L1 norm which tends to force more sparse representations (weights near zero). L2 regularization uses the L2 norm which prevents the weights from getting too large. Dropout is a completely different approach in which neurons are randomly disabled during training to prevent the network from becoming overly reliant on any one unit.

### Creating Our Model
First, let's create a model that overfits. Let's use two hidden layers, each containing 64 nodes. We will use He initialization on each layer and RMSProp as our optimizer. We will train it for 100 epochs to give it ample time to fit to the training data. 

```python
model_overfit = models.Sequential()

# Architecture: Two, 64-node layers plus a softmax activation
model_overfit.add(layers.Dense(64, activation='relu', kernel_initializer='he_uniform', input_shape=(x_train[0].shape)))
model_overfit.add(layers.Dense(64, activation='relu', kernel_initializer='he_uniform'))
model_overfit.add(layers.Dense(10, activation='softmax', kernel_initializer='he_uniform'))

# Compilation: Use RMSProp 
model_overfit.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

# Train for 100 epochs with batch size 256
model_overfit_history = model_overfit.fit(x_train, y_train, epochs=100, batch_size=256, validation_data=(x_val, y_val), verbose=0)
```

Let's write a helper for displaying model performance and see how we did.

```python
def summarize_performance(history):
    display_loss(history)
    display_accuracy(history)
    print('Training Accuracy', history.history['acc'][-1])
    print('Validation Accuracy', history.history['val_acc'][-1])
    
summarize_performance(model_overfit_history)
```

As we can see, there is pretty significant overfitting in the model. The training and validation accuracy are nearly 10% apart, and it appears that our validation accuracy stopped improving as early as epoch 15.

### L1 Regularization
We add L1 regularization with the `kernel_regularizer` parameter. The 0.003 is a hyperparameter which tells us how much the L1 norm is multiplied by. The larger this number is, the larger the regularization effect will be.

```python
from keras import regularizers

# Architecture: Add L1 regularizer to first two hidden layers
model_L1 = models.Sequential()
model_L1.add(layers.Dense(64, activation='relu', kernel_initializer='he_uniform', kernel_regularizer=regularizers.l1(0.003), input_shape=(x_train[0].shape)))
model_L1.add(layers.Dense(64, activation='relu', kernel_initializer='he_uniform', kernel_regularizer=regularizers.l1(0.003),))
model_L1.add(layers.Dense(10, activation='softmax', kernel_initializer='he_uniform'))

# Compilation
model_L1.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

# Train
model_L1_history = model_L1.fit(x_train, y_train, epochs=250, batch_size=256, validation_data=(x_val, y_val), verbose=0)

# Summarize
summarize_performance(model_L1_history)
```

### L2 Regularization
The L2 regularization is used the same was as the L1 regularizer.

```python
# Architecture: Add L1 regularizer to first two hidden layers
model_L2 = models.Sequential()
model_L2.add(layers.Dense(64, activation='relu', kernel_initializer='he_uniform', kernel_regularizer=regularizers.l2(0.003), input_shape=(x_train[0].shape)))
model_L2.add(layers.Dense(64, activation='relu', kernel_initializer='he_uniform', kernel_regularizer=regularizers.l2(0.003),))
model_L2.add(layers.Dense(10, activation='softmax', kernel_initializer='he_uniform'))

# Compilation
model_L2.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

# Train
model_L2_history = model_L2.fit(x_train, y_train, epochs=250, batch_size=256, validation_data=(x_val, y_val), verbose=0)

# Summarize
summarize_performance(model_L2_history)
```

### Dropout
Dropout is completely different from L1 and L2 regularization. It is actually a layer in the network! The number 0.3 is a hyperparameter that tells us what percentage of units to randomly disable during training. In this case, 30% of units in the previous layer will be randomly deactivated during each iteration. 

```python
# Architecure: Add dropout to the two hidden layers
model_drop = models.Sequential()
model_drop.add(layers.Dense(64, activation='relu', kernel_initializer='he_uniform', input_shape=(x_train[0].shape)))
model_drop.add(layers.Dropout(0.3))
model_drop.add(layers.Dense(64, activation='relu', kernel_initializer='he_uniform'))
model_drop.add(layers.Dropout(0.3))
model_drop.add(layers.Dense(10, activation='softmax', kernel_initializer='he_uniform'))

# Compilation 
model_drop.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

# Train
model_drop_history = model_drop.fit(x_train, y_train, epochs=100, batch_size=256, validation_data=(x_val, y_val), verbose=0)

# Summarize 
summarize_performance(model_drop_history)
```

### Summary 
The results are summarized below. As you can see, regularization helps keep the training and validation accuracies a lot closer. It did, however, result in lower training accuracies, so we could try to train for more epochs or attempt to improve our model architecture and see if we can improve the results. Note that regularization is for combatting overfitting, so if you aren't even getting good training accuracy, it's not time to add regularization yet.

| Dataset    | L1    | L2    | Dropout |
